# Heart Disease Analysis

```python
#Importing necessary data analysis librarie
%matplotlib inline 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing heart disease data set
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
# Plotting histogram on the basis of some cols
heart_disease["age"].plot.hist(bins=40);
```

```python
heart_disease["thalach"].plot.hist(bins=40);
```

```python
# Plotting all the cols subplots
heart_disease.plot.hist(figsize=(10,25), subplots=True);
```

```python
# Perform data analysis on patients over 40
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(over_50)
```

### Making a final Heart Disease Analysis graph

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1, 
                               sharex=True, 
                               figsize=(10,10))


# Axis[0,1](row 0, column 1)
scatter = ax0.scatter(over_40["age"],
                     over_40["chol"],
                     c=over_40["target"],
                     cmap='winter')


# Customizing the figure
ax0.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")


# Set the X axis
ax0.set_xlim([50,80])


# Setup a mean line
ax0.axhline(y=over_40["chol"].mean(),
           color='r',
           linestyle='--',
           label="Average");

# Because of my matplotlib version conflict, whenever trying to use legend in my figure,
# I have faced some issues, that's why I'm not using this feature right now, but i leave this code for my future usage.
# ax0.legend(*scatter.legend_elements(), title='Target')

# Axis[1,1](row 1, column 1)
scatter = ax1.scatter(over_40["age"], 
                      over_40["thalach"], 
                      c=over_40["target"],
                      cmap='winter')
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Set the y axis
ax1.set_ylim([60, 200])

# Setup a mean line
ax1.axhline(y=over_40["thalach"].mean(), 
            color='r', 
            linestyle='--', 
            label="Average");

# Because of my matplotlib version conflict, whenever trying to use legend in my figure,
# I have faced some issues, that's why I'm not using this feature right now, but i leave this code for my future usage.
# ax1.legend(*scatter.legend_elements(), title="Target")

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
